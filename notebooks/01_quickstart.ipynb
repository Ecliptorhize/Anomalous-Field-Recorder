# AFR Quickstart

Minimal end-to-end example: load a config, simulate acquisition, process, and generate a report.

```python
from pathlib import Path

from anomalous_field_recorder import (
    generate_report,
    load_experiment_config,
    process_dataset,
    simulate_acquisition,
)

BASE_DIR = Path("..").resolve()
CONFIG = BASE_DIR / "docs/experiments/field-baseline.yml"
RAW_DIR = BASE_DIR / "data/raw/notebook-demo"
PROCESSED_DIR = BASE_DIR / "data/processed/notebook-demo"
```

```python
config = load_experiment_config(CONFIG)
config
```

```python
metadata = simulate_acquisition(CONFIG, RAW_DIR)
metadata["domain_profile"]
```

```python
summary = process_dataset(RAW_DIR, PROCESSED_DIR, band=(1, 40), notch=50)
summary
```

```python
report_path = generate_report(PROCESSED_DIR)
report_path
```
