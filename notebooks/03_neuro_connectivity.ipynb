# Computational Neuroscience: Connectivity

Compute coherence and phase-locking value (PLV) between synthetic EEG channels, plus event-locked peaks.

```python
from anomalous_field_recorder import (
    compute_coherence,
    compute_event_locked_peak,
    compute_phase_locking_value,
    generate_multichannel_eeg,
)

channels, events = generate_multichannel_eeg(num_channels=2, duration_s=5.0, sample_rate=256.0)
coh = compute_coherence(channels, sample_rate=256.0)
plv = compute_phase_locking_value(channels)
erp_peak = compute_event_locked_peak(channels[0], events, sample_rate=256.0)
coh, plv, erp_peak
```

The coherence matrix gives average magnitude-squared coherence per channel pair, while PLV summarizes phase synchrony. Event-locked peak reports the average absolute amplitude around the synthetic event markers.
